```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# load datasets
film_perf = pd.read_csv("film_performance.csv")
customer_perf = pd.read_csv("customer_behavior.csv")
monthly_trends = pd.read_csv("monthly_trends.csv")
store_staff = pd.read_csv("store_staff_performance.csv")
geo_market = pd.read_csv("geographic_market.csv")
inventory_ops = pd.read_csv("inventory_operations.csv")
```

```python
datasets = {
    "film_perf": film_perf,
    "customer_perf": customer_perf,
    "monthly_trends": monthly_trends,
    "store_staff": store_staff,
    "geo_market": geo_market,
    "inventory_ops": inventory_ops
}

for name, df in datasets.items():
    print(f"\n{name.upper()}")
    display(df.head())
    print(df.info())
```

```python
# converting date columns to datetime
date_cols = [
    "last_rental_date", "first_rental_date",
    "most_recent_rental_date", "year_month"
]
for col in date_cols:
    for df in datasets.values():
        if col in df.columns:
            df[col] = pd.to_datetime(df[col])
```

```python
main_df = pd.merge(film_perf , inventory_ops, on=["film_id", "film_title", "category", "total_revenue","last_rental_date", "total_rental"],)
main_df.to_csv("film_and_inventory.csv", index=False)
main_df.head()
```
